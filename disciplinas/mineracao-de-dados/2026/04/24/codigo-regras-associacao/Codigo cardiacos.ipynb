```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# 1. Ler base local
df = pd.read_csv('pacientes.csv')
```

```python
# 2. Garantir que todas as colunas sejam categóricas (texto)
df = df.astype(str)
# 3. Codificação one-hot
df_encoded = pd.get_dummies(df)
```

```python
# 4. Aplicar algoritmo Apriori (suporte mínimo de 0.5 = 50%)
frequent_itemsets = apriori(df_encoded, min_support=0.5, use_colnames=True)
```

```python
# 5. Gerar regras de associação (confiança mínima de 0.8)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
```

```python
# 6. Filtrar apenas regras cujo consequente é 'doenca_Sim'
rules_doenca = rules[rules['consequents'].astype(str).str.contains('doenca_Sim')]
```

```python
# 8. Exibir regras completas em formato textual
print("🔹 Regras de Associação (com descrição completa):\n")
for idx, row in rules_doenca.iterrows():
    antecedente = ', '.join(list(row['antecedents']))
    consequente = ', '.join(list(row['consequents']))
    print(f"➡️ SE ({antecedente})  ENTÃO ({consequente})")
    print(f"   Suporte: {row['support']:.2f} | Confiança: {row['confidence']:.2f}\n")
```
